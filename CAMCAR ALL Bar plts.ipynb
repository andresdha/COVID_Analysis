```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [25,20]
import seaborn as sns
sns.set(palette='bright')
```

```python
cases = pd.read_csv('/home/andres/Documents/JupyterNBs/COVID_19_Data/COVID_Analysis/All_CAMCAR_Cases.csv',index_col=0)
deaths = pd.read_csv('/home/andres/Documents/JupyterNBs/COVID_19_Data/COVID_Analysis/All_CAMCAR_Deaths.csv',index_col=0)
recovered = pd.read_csv('/home/andres/Documents/JupyterNBs/COVID_19_Data/COVID_Analysis/All_CAMCAR_Recovered.csv',index_col=0)
```

```python
cases
```

```python
deaths.head()
```

```python
recovered.head()
```

```python
names_cases = cases[cases['Province/State'].notna()].iloc[:,0]
indices = names_cases.index
```

```python
cases.iloc[indices, 1] = names_cases
```

```python
cases.drop('Province/State', axis=1, inplace=True)
```

```python
cases
```

```python
names_deaths = deaths[deaths['Province/State'].notna()].iloc[:,0]
indices = names_deaths.index
deaths.iloc[indices, 1] = names_deaths
deaths.drop('Province/State', axis=1, inplace=True)
deaths
```

```python
names_recovered = recovered[recovered['Province/State'].notna()].iloc[:,0]
indices = names_recovered.index
recovered.iloc[indices, 1] = names_recovered
recovered.drop('Province/State', axis=1, inplace=True)
recovered
```

```python
cases.set_index('Country/Region', inplace=True)
deaths.set_index('Country/Region', inplace=True)
recovered.set_index('Country/Region', inplace=True)
```

```python
cases
```

```python
deaths
```

```python
indices = cases.index
indices
```

```python
deaths = deaths.reindex(indices)
deaths
```

```python
recovered = recovered.reindex(indices)
recovered
```

```python
day = cases.columns.values[-1]
day
```

```python
active = pd.DataFrame(cases - deaths - recovered)
```

```python
active
```

```python
ax = cases[day].plot.barh(color=[1,0,0,1], x='Country/Region', y=day, label='Total Confirmed Cases')
active[day].plot.barh(ax=ax, color=[1,0.9,0,1], label='Active Cases')
deaths[day].plot.barh(ax=ax, color=[0,0,0,1], label='Deaths')
recovered[day].plot.barh(ax=ax, color=[0,1,0,0.5], label='Recovered')
plt.yticks(fontsize=20, rotation=0)
plt.ylabel(day, fontsize=30)
plt.xticks(np.arange(0, cases.iloc[cases[day].argmax(), 1], 25),fontsize=15, rotation=45)
plt.title('COVID-19 CAMCAR', fontsize=25)
plt.legend(loc='best', fontsize=25)
plt.show()
```

```python
joined_data = pd.DataFrame(cases[day])
joined_data
```

```python
joined_data.rename(columns={day: 'Cases'}, inplace=True)
```

```python
joined_data['Deaths'] = deaths[day]
```

```python
joined_data['Recovered'] = recovered[day]
```

```python
joined_data['Active'] = active[day]
```

```python
joined_data.to_csv('All_CAMCAR_Latest_Cases_Deaths_Recovered_Active.csv', index=True)
```

```python
joined_data.plot(kind='barh', cmap=plt.get_cmap('jet'))
plt.yticks(fontsize=20, rotation=0)
plt.ylabel(day, fontsize=30)
plt.xticks(np.arange(0, cases.iloc[cases[day].argmax(), 1], 25),fontsize=15, rotation=45)
plt.title('COVID-19 CAMCAR Data', fontsize=25)
plt.legend(loc='best', fontsize=25)
plt.show()
```

```python
data_bar = joined_data.sort_values(by=['Cases'])
#data_bar = data_bar[cols]
data_bar
```

```python
data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].plot(kind='barh', cmap=plt.get_cmap('seismic'))
plt.yticks(fontsize=20, rotation=0)
plt.ylabel('', fontsize=30)
plt.xticks(np.arange(0, cases.iloc[cases[day].argmax(), 1], 25),fontsize=15, rotation=45)
plt.title('COVID-19 CAMCAR Data   ' + day, fontsize=25)
plt.legend(loc=4, fontsize=25)
columns = data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].shape[1]
rows = data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].shape[0]
ys = np.array(range(rows))
for i in range(rows):
    plt.text(
        x=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,0] + 5, 
        y=ys[i] - 0.25, 
        s=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,0], 
        size=20
    )


for i in range(rows):
    plt.text(
        x=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,1] + 5, 
        y=ys[i] -0.1 if data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,2] != 0 else 0.1, 
        s=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,1], 
        size=18
    )


for i in range(rows):
    plt.text(
        x=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,2] + 10, 
        y=ys[i] , 
        s= data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,2], 
        size=18
    )

for i in range(rows):
    plt.text(
        x=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,3] + 5, 
        y=ys[i] + 0.15, 
        s=data_bar[data_bar['Cases'] > data_bar['Cases'].quantile(0.5)].iloc[i,3], 
        size=20
    )
plt.savefig('/home/andres/Documents/JupyterNBs/COVID_19_Data/Figures/CAMCAR_Largest.png')
plt.show()
```
