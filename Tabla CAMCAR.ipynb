```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker
plt.rcParams['figure.figsize'] = [20,15]
import seaborn as sns
sns.set()
sns.set_palette('bright')
```

```python
df = pd.read_csv('All_CAMCAR_Latest_Cases_Deaths_Recovered_Active.csv', index_col=0)
```

```python
df
```

```python
data = df.copy()
```

```python
data['Mortality Ratio'] = data['Deaths']/data['Cases']
```

```python
data['Mortality pct'] = data['Mortality Ratio']*100
```

```python
data
```

```python
pops = [11108011, 4291013, 11495023, 5044015, 9721007, 17907008, 6480013, 2911015, 1376715, 106521, 297083, 287425, 65565, 42777, 405714, 11374286, 38531, 97765, 181214, 163527, 6416009]
```

```python
data
```

```python
data['Population'] = pops
data.loc['Dominican Republic', 'Population'] = pops[0]
data.loc['Panama', 'Population'] = pops[1]
data.loc['Honduras', 'Population'] = pops[4]
data.loc['Cuba', 'Population'] = pops[2]
data.loc['Guatemala', 'Population'] = pops[5]
data.loc['El Salvador', 'Population'] = pops[6]
data.loc['Costa Rica', 'Population'] = pops[3]
data.loc['Jamaica', 'Population'] = pops[7]
data.loc['Trinidad and Tobago', 'Population'] = pops[8]
data.loc['Aruba', 'Population'] = pops[9]
data.loc['Bahamas', 'Population'] = pops[14]
data.loc['Barbados', 'Population'] = pops[11]
data.loc['Nicaragua', 'Population'] = pops[-1]
data.loc['Saint Lucia', 'Population'] = pops[-3]
data.loc['Curacao', 'Population'] = pops[-2]
```

```python
data.columns.values
```

##### Rate of infection 

Defined as: $$\frac{Confirmed}{Population/100000}$$

```python
data['Rate of Infection'] = data['Cases']/(data['Population']/100000)
```

```python
data
```

#### Daily Incidence pct

$$\frac{New Cases}{Active}*100$$

```python
data['Daily Incidence pct'] = data['New Cases']/data['Active']*100
data
```

```python
data.columns.values
```

```python
cols = ['Population', 'Cases', 'Deaths', 'Recovered', 'Active', 'New Cases',
       'Mortality Ratio', 'Mortality pct', 
       'Rate of Infection', 'Daily Incidence pct']
```

```python
temp = data[cols]
temp
```

```python
temp.drop(['French Guiana', 'Cayman Islands', 'Sint Maarten', 'Haiti', 'St Martin', 'Antigua and Barbuda'], axis=0, inplace=True)
```

```python
result = temp.copy()
```

```python
result
```
